This is a simple Python program that utilises librosa library to detect the bpm of given sample audios and categorises the audios based on the bpm result.

```python
import librosa #loads librosa library into the colab notebook so that we can use its fuctions
import numpy as np #loads numpy library that handles numbers and arrays
```

```python
filename = librosa.ex('nutcracker') #fetch a built in sample audio
audio, sample_rate = librosa.load(filename) #reads samole file and converts it into data that Python can work with
```

```python
tempo, beat_frames = librosa.beat.beat_track(y=audio, sr=sample_rate) #this analyses the rythym of the audio and estimates the tempo
bpm = round(float(tempo[0])) #tempo converted from numpy array into a plain decimal number that is rounded off.
print(f"Detected BPM: {bpm}")
```

```python
def classify_tempo(bpm):
    if bpm < 80:
        return "Slow / Ballad — Lo-fi, Gospel, Slow Jams"
    elif bpm < 120:
        return "Pop / R&B — Afropop, R&B, Reggae"
    elif bpm < 150:
        return "Dance / Hip-hop — Hip-hop, Dancehall, House"
    else:
        return "EDM / High Energy — EDM, Drum & Bass, Amapiano"

category = classify_tempo(bpm)
print(f"Genre Category: {category}")
#function that categorises the sample audio based on the bpm result
```

```python
#print the output from the result obtained giving all the necessary details
print("  BPM Classifier Result")
print(f"  Audio file   : nutcracker")
print(f"  Detected BPM : {bpm}")
print(f"  Genre        : {category}")
```

```python
#this is the second sample test
filename2 = librosa.ex('trumpet')
audio2, sample_rate2 = librosa.load(filename2)
tempo2, _ = librosa.beat.beat_track(y=audio2, sr=sample_rate2)
bpm2 = round(float(tempo[0]))
category2 = classify_tempo(bpm2)


print("  BPM Classifier Result")
print(f"  Audio file   : trumpet")
print(f"  Detected BPM : {bpm2}")
print(f"  Genre        : {category2}")
```
